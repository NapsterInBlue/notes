## Create Integer Variable

```scala
// Create an integer variable of 10
var i:Int = 10
```

## Increment And Decrement

```scala
// Increment up 1
i += 1

// View variable
i
```

```scala
// Decrement up 1
i -= 1

// View variable
i
```

```scala
// Increment up x2
i *= 2

// View variable
i
```

```scala
// Decrement down by half
i /= 2

// View variable
i
```
