## Preliminaries

```python
from collections import deque
```

## Make A Queue

```python
# Option 1: Make a queue
queue = deque(range(10))

# Option 2: Make a queue that, if full, discards any item at the 
# opposite end to where you added an item.
queue = deque(range(10), maxlen=10)
```

## Manipulate Queue

```python
# Append an item to the right

queue.append('A')

# View queue
queue
```

```python
# Append an item to the left

queue.appendleft('A')

# View queue
queue
```

```python
# Count occurances of item
queue.count('A')

# View queue
queue
```

```python
# Remove and return right-most item
queue.pop()

# View queue
queue
```

```python
# Remove and return left-most item
queue.popleft()


# View queue
queue
```

```python
# Insert item to the right of an item
queue.insert(2, 'A')

# View queue
queue
```

```python
# Reverse the queue
queue.reverse()

# View queue
queue
```

```python
# Delete all items
queue.clear()

# View queue
queue
```
