## Preliminaries

## Load Image As Greyscale

```python
# Load image as grayscale
image = cv2.imread('images/plane_256x256.jpg', cv2.IMREAD_GRAYSCALE)
```

## Crop Image

```python
# Select first half of the columns and all rows
image_cropped = image[:,:126]
```

## View Image
